```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate
```

```python
# Parameters
sampling_rate = 1000  # Hz
duration = 1  # seconds
frequency = 50  # Hz
noise_level = 0.5  # Standard deviation of noise

# Generate clean sine wave
t = np.linspace(0, duration, sampling_rate * duration)
clean_signal = np.sin(2 * np.pi * frequency * t)

# Generate noisy sine wave
noisy_signal = clean_signal + np.random.normal(0, noise_level, size=len(clean_signal))

# Perform autocorrelation
correlation = correlate(noisy_signal, clean_signal, mode='full')

# Find the index of the maximum correlation
max_index = np.argmax(correlation)

# Calculate the time of flight
time_of_flight = (max_index - len(clean_signal) + 1) / sampling_rate

# Plot the signals and autocorrelation
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, clean_signal)
plt.title('Clean Sine Wave')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, noisy_signal)
plt.title('Noisy Sine Wave')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(np.arange(-len(clean_signal) + 1, len(clean_signal)), correlation)
plt.title('Autocorrelation')
plt.xlabel('Lag (samples)')
plt.ylabel('Correlation')

plt.tight_layout()
plt.show()

print("Time of flight:", time_of_flight, "seconds")
```
